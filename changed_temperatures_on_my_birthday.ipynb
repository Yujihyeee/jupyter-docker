```python
import random
import matplotlib.pyplot as plt
```

```python
import csv
```

```python
data = csv.reader(open('data/seoul.csv', 'rt', encoding='UTF-8'))
```

```python
next(data)
```

```python
ls = list(data)
```

```python
print([i for i in ls])
```

```python
print([i[-1] for i in ls]) # show_highest_temperature
```

```python
highest_temperatures = []
[highest_temperatures.append(float(i[-1])) for i in ls if i[-1] != '']
print(f'총 {len(highest_temperatures)} 개')
```

```python
plt.figure(figsize=(20, 2))
plt.plot(highest_temperatures, 'r')
```

```python
high = []  # 최고기온
low = []  # 최저기온
```

```python
for i in ls:
    if i[-1] != '' and i[-2] != '':
        if 1995 <= int(i[0].split('-')[0]):
            if i[0].split('-')[1] == '05' and i[0].split('-')[2] == '18':
                high.append(float(i[-1]))
                low.append(float(i[-2]))
```

```python
plt.rc('font')
plt.rcParams['axes.unicode_minus'] = False
plt.title('temperature_on_my_birthday')
plt.plot(high, 'hotpink', label='high')
plt.plot(low, 'skyblue', label='low')
plt.legend()
```
